```python
x1 = np.array(x1)
x2 = np.array(x2)
x3 = np.array(x3)
y = np.array(y)
```

```python
n = len(x1)
n
```

```python
x_bias = np.ones((n,1))
```

```python
x1_new = np.reshape(x1, (n,1))
x2_new = np.reshape(x2, (n,1))
x3_new = np.reshape(x3, (n,1))
```

```python
x_new = np.append(x_bias, x1_new, axis = 1)
x_new = np.append(x_new, x2_new, axis = 1)
x_new = np.append(x_new, x3_new, axis = 1)
```

```python
x_new
```

```python
x_new_t = np.transpose(x_new)
```

```python
x_new_t_dot_x_new = x_new_t.dot(x_new)
```

```python
temp1 = np.linalg.inv(x_new_t_dot_x_new)
```

```python
temp2 = x_new_t.dot(y)
```

```python
theta = temp1.dot(temp2)
theta
```

```python
def predict(b, w1, w2, w3, bhk, size, bathroom):
  predicted_val = b + w1*bhk + w2*size + w3*bathroom
  return predicted_val
```

```python
print(predict(theta[0],theta[1],theta[2],theta[3],2,800,1))
```
